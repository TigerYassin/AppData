## setting up variables 

```python
import pandas as pd

text_path = 'AppleStore.csv'
df = pd.read_csv(text_path)
df.head()
```

```python
df.sort_values('rating_count_ver', ascending=False).head()
```

```python
groups = df.groupby("prime_genre") #returns a groupby Dataframe
# how to iterate through the groupby dataframe 
'''
for g1, g2 in groups:   
    print(g1)
    print(g2)
'''

lifestyle = groups.get_group("Lifestyle") #another way to access groups in the dataframe 
lifestyle.head() #will only output apps from "lifestyle"
```
